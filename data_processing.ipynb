```python
import os
import pandas as pd
```

```python
csv_dir = 'Data/2024_stats/CSV/'
```

```python
csv_files = [os.path.join(csv_dir, file) for file in os.listdir(csv_dir) if file.endswith('.csv')]
```

```python
merged_df = None
```

```python
for file in csv_files:
    df = pd.read_csv(file)
    df = df.rename(columns=lambda x: x.strip())
    
    df = df.drop(columns=[col for col in ['Rank', 'GM', 'W-L'] if col in df.columns], errors='ignore')

    if merged_df is None:
        merged_df = df  # Set the first dataframe
    else:
        merged_df = pd.merge(merged_df, df, on="Team", how="outer")  # Merge on "Team"
        print(file)
```

```python
merged_df.to_csv('Data/2024_stats/2024_merged_data.csv', index=False)
```
